## 说明
这个模块用来记录常见的激活函数（不含可训练参数的那种）及其基本属性，目前记录了如下几个激活函数:
- relu
- softplus
- sigmoid
- tanh
- swish

```python
import numpy as np
import matplotlib.pyplot as plt
```

## relu函数
- 函数输出值的范围是[0, +∞)， 数学表达式如下：
$$
relu(x) = max(0, x)
$$
- 优点
    - 计算复杂度低，没有指数运算
    - 是神经网络中使用最多的基础激活函数（没有之一）
    - 收敛速度快，不会造成梯度爆炸的问题
- 缺点
    - 非零均值函数
    - 会造成dead relu problem即神经元坏死的状况（负数部分的梯度都为0，永远不会被更新）

```python
x = np.arange(-10, 10, 0.1)
y = [max(0,i) for i in x]
plt.plot(x, y)
plt.show()
```

## softplus函数
值域范围是(0,+∞),其数学表达式如下：
$$
softplus(x) = log(1+e^x)
$$
与relu相比不会发生神经元坏死的情况

```python
def softplus(x):
    return np.log(1+np.exp(x))
x = np.arange(-10, 10, 0.1)
y = softplus(x)
plt.plot(x, y)
plt.show()
```

## sigmoid函数
- 是一个"S"型函数，函数输出值的范围是(0, 1)，其数学表达式如下:
$$
sigmoid(x) = \frac{1}{1+e^{-x}}
$$
- 优点
    - 能够对信息进行压缩
    - 能够把输出归约到0到1之间，可以把回归的结果转化成概率展示
- 缺点
    - 容易发生梯度消失问题，导致神经网络无法收敛或者收敛十分困难
    - 有一定的概率发生梯度爆炸的问题，特别是$x=0$附近的梯度值
    - 计算相对relu来说更加的耗时，涉及到了幂运算
    - 非零均值函数

```python
def sigmoid(x):
    '''
    x是一个np.array对象
    '''
    return 1/(1+np.exp(-x))
x = np.arange(-10, 10, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

## tanh函数
- tanh函数与sigmoid函数类似，都是一个"S"型函数，可以说tanh是sigmoid的一个延伸，它的数学如下所示:
$$
tanh(x) = 2\cdot sigmoid(2x)-1 = \frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}
$$
- 与sigmoid不同的是，tanh函数是零均值函数，其输出范围是(-1, 1)。
- 与sigmoid相同，同样存在资源消耗多，容易导致梯度消失和梯度爆炸的问题。

```python
def tanh(x):
    '''
    x是一个np.array对象
    '''
    return 2*sigmoid(x)-1
x = np.arange(-10, 10, 0.1)
y = tanh(x)
plt.plot(x, y)
plt.show()
```

## swish函数
swish函数是由google使用搜索技术得到的一个激活函数，号称在大多数情况下都能够比relu函数有更好的表现结果，其数学表达式如下：
$$
swish(x) = x\cdot sigmoid(x)
$$


```python
def swish(x):
    return x*sigmoid(x)
x = np.arange(-10, 10, 0.1)
y = swish(x)
plt.plot(x, y)
plt.show()
```

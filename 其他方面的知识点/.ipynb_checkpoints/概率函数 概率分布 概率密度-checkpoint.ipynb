## 相关内容
- 积分：根据导数求取原函数
- 微分：求取已知函数的导数
- 不定积分：指积分后得到的原函数常数项无法确定
- 定积分：被积式有上下限，得到的是一个数值不是一个函数

## 简介
三个东西都是用来描述概率的函数，本文重点记录连续性随机变量的相关知识。

## 概率函数
- 对于离散变量来说，我们可以轻易的得到取某个值的概率是多少，如果函数的输入是某个值输出是概率，那么我们就称这个函数叫做概率函数，其基本形式可以表示成这样：
$$
f(x) = p(X=x)
$$

- 对于连续型变量来说，是没有办法通过上述的这个概率函数（传入某个值）来得到概率的，因为连续型取某个值的概率接近于0

## 概率分布
- 概率分布所描述的是**所有可能概率的分布情况**（缺一个可能的概率都不可）
- 概率分布函数又称为累计概率函数，数学表达式如下:
$$
F(x) = p(X \leq x) = \sum_{x_k \leq x} p_k
$$
- 对于连续型变量来说，事件$\{x\lt X \leq x+h\}$的概率用概率分布函数来表示的话应该是$F(x+h) - F(x)$

## 概率密度
- 概率密度函数就是**概率分布函数的导函数即微分**，概率密度函数的$y$值就是函数概率的变化率
- 概率密度函数的输出值很有可能大于1，概率密度函数与横轴所包围的面积不一定等于1，其面积值只能说明概率的一种分布。
- **概率密度值不等于概率**，应该是概率在该点的变化。

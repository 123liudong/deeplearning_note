## 联合概率
表示多个事件同时发生的概率，如事件A和事件B同时发生：
$$
p(A, B) = P(A \cap B)
$$

## 条件概率
表示在某些条件发生后，某一事件发生的概率，如事件A已经发生，求事件B发生的概率：
$$
p(B|A) = \frac{p(B, A)}{p(A)}
$$

## 边缘概率
指定义在子集上的概率，这个概率求和不一定等于1，大多数情况都是小于1的。
假设现在有两个维度$x和y$,$x$的取值范围有3个值（1、2、3），$y$的取值范围有6个值（4、5、6、7、8、9），这些概率可以表示为一个3行6列的矩阵。
其中第一行就叫做关于$x=1$的边缘概率分布.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 高斯分布(常态分布/正态分布)
- 正态分布是一个呈钟形的曲线，两头低、中间高且左右对称。
- 该分布的概率密度函数数学表达式如下：
$$
f(x) = \frac{1}{\sqrt{2\pi}\sigma}exp(-\frac{(x-\mu)^2}{2\sigma^2})
$$
当均值为0，方差为1就被称为标准正态分布，将其他分布转为标准正态分布的公式如下：
$$
X \sim N(\mu, \sigma)，Y = \frac{X-\mu}{\sigma} \sim N(0, 1)
$$
其中$\mu$表示均值（也叫数学期望）、$\sigma ^2$是方差
- $\mu$在坐标轴上表现为左右的对称轴，$\sigma$越大曲线越肥

```python
def gaussian_f(x, u, sigma):
    result = 1/(np.sqrt(2*np.pi)*sigma)*np.exp(-np.power(x-u, 2)/(2*np.power(sigma, 2)))
    return result
x = np.arange(-100, 100, 0.1)
y = gaussian_f(x,0, 10)
plt.plot(x, y)
plt.show()
```

## 拉普拉斯分布
其概率密度函数的数学表达式如下：
$$
f(x|u, b) = \frac{1}{2b}exp(-\frac{|x-\mu|}{b}) 
$$
- 其中$\mu$是位置参数，与正态分布的$\mu$类似。
- $b>0$是尺度参数，越大函数越肥

```python
def laplace_f(x, u, b):
    result = np.exp(-np.abs(x-u)/b)/(2*b)
    return result
x = np.arange(-100, 100, 0.1)
y = laplace_f(x,0, 20)
plt.plot(x, y)
plt.show()
```

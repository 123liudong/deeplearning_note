## 为什么要对数值类型的特征做归一化？
1. 为了消除数据特征之间的量纲影响，我们需要对特征进行归一化，使不同指标之间具有可比性即让各指标处于同一数值量级，以便进行分析。
2. 数值归一化后，模型会收敛的更快。
3. 一般通过梯度下降法求解最优结果的模型都需要归一化。


## 高维特征导致的一些问题
- 进行相似性度量比较困难，很难得到有效的衡量。
- 模型参数会随着维度的增加而增加，容易导致过拟合问题。
- 通常只有部分维度对结果有帮助。
- 比较耗资源，主要是内存和计算资源。

## 常见的归一化方法
- 线性归一化（最大最小归一化方法/Min-Max Scaling）,对数据进行等比缩放
$$
X_{norm} = \frac{X - X_{min}}{X_{max}-X_{min}}
$$


- 零均值归一化（Z-Score）,把数据归一化到均值为0、标准差为1的分布上
$$
X_{norm} = \frac{x-\mu}{\sigma}
$$

## 类别型特征/离散型特征进行编码
对于大多数模型来说，类别数据都必须转化成数值才能做进一步处理，比如自然语言处理需要把单词转化成一个数字或者用one-hot来表示。
- 序号编码，适用于类别间具有大小关系的数据。
> eg. (大、中、小)；（低、中、高）

- one-hot编码，一般用来处理类别之间没有级别关系的特征，即类与类之间是平等的关系。但是one-hot编码并不是万能的，他还有这些缺点：
    - 当类别数量过多，成千上万时会导致维数灾难。
    - one-hot太浪费内存了，因为在所有位置上只有1个值具有信息。
    - 一般需要配合特征来降低维度。
   
> eg.(成功，失败)；（A型血、B型血、AB型血、O型血）

- 二进制编码，个人认为是one-hot编码的改进，能有效的降低维度，节省存储空间（信息的提取是否会变得更加困难？）

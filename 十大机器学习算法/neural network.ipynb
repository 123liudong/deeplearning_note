## 神经网络算法
神经网络算法是一个黑盒算法，早在19世纪六七十年代就被提出，近年来随着硬件的发展开始大放光芒，个人认为算是开启了人工智能时代，吹响了第一声号角。

神经网络算法的拟合能力很强，有大牛已经证明神经网络在理论上可以拟合任何函数，因其强大的数据挖掘能力和表达能力，因此被大规模使用。

神经网络个人将之分为以下几种（其他更复杂的神经网络都是由这些基础的神经网络所组成的）：
1. 全连接神经网络
2. 卷积神经网络
3. 循环神经网络

## 神经网络基础
1. 矩阵的乘法(加权求和或者其他的方式)
2. 非线性转换规则（激活函数）


- 假设我们现在两个矩阵，形状分别是(1, m), (m, n)，通过矩阵的点积我们可以得到形状是(1, n)的矩阵,如下所示:
$$
(1, m)\cdot (m, n) = (1, n)
$$
这是一个很简单的矩阵相乘，前一个矩阵的行与后一个矩阵的列相乘然后求和得到一个数，很明显这是一个**线性的操作**，这个操作过程是全连接神经网络中的基础步骤之一

- 然而神经网络是一个非线性的模型，上面的操作不具备任何非线性因素，所以我们还需要对形状为(1, n)的矩阵做一个非线性转换，假设这个非线性转换叫做函数$af(x)$，其转换过程如下所示：
$$
 \left[
 \begin{matrix}
   af(x_{11}) & af(x_{12}) & af(x_{13}) & ... & af(x_{1n}) 
  \end{matrix}
  \right] \tag{3}
$$

- 这样我们得到的就是一个经过非线性转换函数$af(x)$转换过的矩阵了，我们得到这样的一个矩阵数据，这个矩阵可以作为下一个矩阵相乘的前一个矩阵，然后重复这两个过程就是一个最简单的神经网络。

- 矩阵相乘表示加权求和，非线性转换我们称之为激活函数，非线性转换的时候矩阵形状不变。

- 假设我们的神经网络是一个回归模型，有一条数据用向量来表示，形状是(1, m)，神经网络层数有 4 层，每层的神经元个数分别是(10, 20, 30， 1)，作为模型参数的矩阵形状依次是(m, 10)、(10, 20)、(20, 30)、(30, 1)，其形状变化过程如下:
$$
(1, m) \cdot (m, 10)\cdot (10, 20)\cdot (20, 30)\cdot (30, 1)=(1, 1)
$$
简单的数学公式表示如下:
$$
tmp1 = af_1(m_{(1, m)}\cdot m_{(m, 10)})\\
tmp2 = af_2(tmp1\cdot m_{(m, 20)})\\
tmp3 = af_3(tmp2\cdot m_{(m, 30)})\\
tmp4 = af_4(tmp3\cdot m_{(m, 10)})\\
$$
